Question 2

Join the data containing card details and customer details by customer id, so that all records of card details and any matching records in customer details are kept.

```python
card_cust= card_base.merge(customer_base,how = "left", on = "Cust_ID")
card_cust
```

Question 3

Join the data containing fraud details with transaction details so all rows of both tables are kept. What does the resulting row number tell you?

```python
fraud_trans = fraud_base.merge(transaction_base, how = "outer", on = "Transaction_ID")
fraud_trans
```

By using an outer join we have tacked on fraud flags for those transactions that are applicable while keeping all transactions

Question 4

[Harder] Join the data containing card details with transaction details so rows from the first which have matching ones in the second are returned.

Hint: semi is not an allowed option for how in .merge(). You'll need to think about the process of a semi join and then perform each step.

```python
full_fraud = card_cust.merge(fraud_trans, how = "inner", left_on = "Card_Number", right_on = "Credit_Card_ID")
full_fraud = full_fraud.loc[~full_fraud["Fraud_Flag"].isna()]
full_fraud
```

```python
hats = pd.read_csv("data/hat_observations.csv")
hats_split= hats["observation"].str.split(",", expand = True)
hats_split
```

```python
hats.insert(loc = 5, column = "colour", value = hats_split[0])
hats.insert(loc = 6, column = "style", value = hats_split[1])
hats.drop("observation", axis = 1, inplace = True)
hats
```
